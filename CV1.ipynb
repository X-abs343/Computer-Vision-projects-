```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.io import imshow
from skimage.color import rgb2gray
```

```python
shape1 = np.zeros((2,2))
for i in range(2):
    for j in range(2):
        if (i + j) % 2 == 1:
            shape1[i, j] = 255
plt.imshow(shape1, cmap='gray', vmin=0, vmax=255)
```

```python
shape2 = np.zeros((2,2))
for i in range(2):
    for j in range(2):
        if (i + j) % 2 == 0:
            shape2[i, j] = 255
plt.imshow(shape2, cmap='gray', vmin=0, vmax=255)
```

```python
result_matrix = np.clip(shape1 + shape2, 0, 255)
result_matrix
```

```python
chessboard = np.zeros((8,8))
for i in range(8):
    for j in range(8):
        if (i + j) % 2 == 1:
            chessboard[i, j] = 255
plt.imshow(chessboard, cmap='gray', vmin=0, vmax=255)
```

```python
chessboard = np.zeros((1024,1024))
for i in range(1024):
    for j in range(1024):
        if (i + j) >= 1024:
            chessboard[i, j] = 255
        else:
          chessboard[i, j] = 0
plt.imshow(chessboard, cmap='gray', vmin=0, vmax=255)
```
