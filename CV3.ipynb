```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.io import imshow
```

```python
image1= cv2.imread('IMG1.png')
plt.imshow(image1)
```

```python
image1
```

```python
ret, mask = cv2.threshold(image1,50,255,cv2.THRESH_BINARY)
plt.imshow(mask)
kernal = np.ones((5,5),np.uint8) #matrix size 5
```

```python
dilation= cv2.dilate(mask,kernal,iterations=2)
plt.imshow(dilation)
```

```python
erosion = cv2.erode(mask,kernal, iterations=7)
plt.imshow(erosion)
```

```python
dilation2= cv2.dilate(erosion,kernal,iterations=2)
plt.imshow(dilation2)
```

```python
image2= cv2.imread('IMG2.png')
plt.imshow(image2)
```

```python
ret, mask1 = cv2.threshold(image2,100,255,cv2.THRESH_BINARY)
plt.imshow(mask1)
```

```python
#erosion followed by dilation
opening= cv2.morphologyEx(mask1,cv2.MORPH_OPEN,kernal)
plt.imshow(opening)
```
